```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('data/allergy.csv')
df.info()
df.head()
```

```python
# supprimer les colonnes inutiles
df=df.drop(['allergyid','patientunitstayid','drughiclseqno','drugname'],axis=1)
df.info()
df.head()
```

```python
# supprimer les lignes avec des valeurs manquantes
df=df.dropna()
df.info()
```

```python
dummies=[]
cols =["rxincluded","writtenineicu","specialtytype"]
for col in cols:
    dummies.append(pd.get_dummies(df[col], prefix=col))
dummies_df=pd.concat(dummies,axis=1)
df=pd.concat((df,dummies_df),axis=1)
df=df.drop(cols,axis=1)
df.info()
```

```python
# diviser les données en ensembles d'entraînement et de test
X=df.drop('allergytype',axis=1)
y=df['allergytype']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```
