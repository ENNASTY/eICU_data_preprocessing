```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('data/medication.csv')
df.info()
df.head()
```

```python
#Supprimez les colonnes qui ne sont pas utiles
cols=['medicationid','patientunitstayid','loadingdose','gtc','drugname','drughiclseqno','dosage']
df=df.drop(cols,axis=1)
df.info()
df.head()
```

```python
#Supprimez les lignes avec des valeurs manquantes
df=df.dropna()
df.info()
df.head()
```

```python
dummies = []
cols = ['drugivadmixture', 'prn']
for col in cols:
    dummies.append(pd.get_dummies(df[col], prefix=col))

dummies_df = pd.concat(dummies, axis=1)
df = pd.concat([df, dummies_df], axis=1)
df = df.drop(cols, axis=1)
df.info()
```

```python
X= df.values
y= df["drugordercancelled"].values
X= np.delete(X, 3, axis=1)

X_train, X_test, Y_train, Y_test= train_test_split(X, y, test_size=0.3, random_state=0)
```
